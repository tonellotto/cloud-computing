# Virtual Machines Setup
----
This notebook contains the preliminary steps necessary to set up and configure correctly the network and the software of our virtual machines.
In particular:

0. We will [retrieve the IP addresses](#preliminaries) of our virtual machines and assign them new hostnames we will use later.
1. We will [create a new user](#creation) called `hadoop`; we will install and use Hadoop as this user.
2. We will [configure a password-less SSH access](#remote_access) among your virtual machines.
4. We will [install the required software](#install) on each virtual machines.
5. We will [configure the network](#configure) of your virtual machines.

**Each step assumes that you start with a open Bash shell on your local machine**.

## 1. Hadoop user creation <a name="creation"/>
----
**On each virtual machine** in your cluster, execute the following steps, one by one, line by line.
<span style="color:red">*Do not copy-paste more than one line at a time*.</span>

Note: When you replace the `<ip address>` field, **always use IP addresses, never use hostnames**.

1. Login as `root` providing the `root` user password when requested.

 ```bash
 ssh root@<ip address>
 ```  

2. Create the `hadoop` user account. Provide a new password for the new `hadoop` user when requested.

 ```bash
 sudo adduser --gecos '' hadoop
 sudo adduser hadoop sudo
 ```  

3. Logout.

 ```bash
 exit
 ```

Repeat the previous step for every virtual machines in your cluster.

## 2. Configure SSH password-less access <a name="remote_access"/>
----
**On every virtual machine** in your cluster, execute the following steps, one by one, line by line. 
<span style="color:red">*Do not copy-paste more than one line at a time*.</span>

Note: When you replace the `<ip address>` field, <span style="color:red">**always use IP addresses, never use hostnames.**</span>

1. Login as `hadoop` providing the `hadoop` user password when requested.

  ```bash
  ssh hadoop@<ip address>
  ```  

2. Create the `.ssh` folder.

  ```bash
  mkdir ~/.ssh
  chmod 700 ~/.ssh
  ```

3. Generate a SSH key-value pair.

  ```bash
  ssh-keygen -t rsa -N '' -f ~/.ssh/id_rsa
  cat ~/.ssh/id_rsa.pub >> ~/.ssh/authorized_keys
  chmod 700 ~/.ssh/authorized_keys
  ```

4. Configure the SSH access.

  ```bash
  echo "StrictHostKeyChecking no" > ~/.ssh/config
  chmod 644 ~/.ssh/config
  ```

5. Check that you can ssh **locally, i.e., from the virtual machine to the virtual machine,** without a passphrase. 

  ```bash
  ssh localhost
  exit
  ```

6. Logout.

  ```bash
  exit
  ```

<span style="color:red">Repeat the previous step for every virtual machines in your cluster.</span>

## 3. Install required software <a name="install"/>
----
**On each virtual machine** in your cluster, execute the following steps, one by one, line by line.
<span style="color:red">*Do not copy-paste more than one line at a time*.</span>


Note: When you replace the `<ip address>` field, <span style="color:red">**always use IP addresses, never use hostnames.**</span>

1. Login as `hadoop`.

    ```bash
    ssh hadoop@<ip address>
    ```  
    
    If asked for a password, you did not complete steps 2 and/or 3 successfully. Go back and repeat.

2. Run the following commands, providing the `root` user password if requested. Some commands can take a lot of time.

    ```bash
    sudo apt update
    sudo apt upgrade -y
    sudo apt install -y nano python3 python3-pip ipython3 openjdk-8-jdk
    sudo apt autoremove -y --purge
    ```

3. Run the following commands, providing the `root` user password if requested, to create the folder we will use:

    ```bash
    sudo mkdir -p /opt/{hadoop/logs,hdfs/{datanode,namenode},yarn/logs,spark/logs}
    sudo chown -R hadoop /opt
    ```

    We will install all our software under the `/opt` directory and store HDFS underlying data there as well.
    The layout of the `/opt` directory will look like:
    
    ```bash
    /opt
     ├── hadoop
     │   └── logs
     ├── hdfs
     │   ├── datanode
     │   └── namenode
     ├── yarn
     │   └── logs
     └── spark
         └── logs
    ```
    
4. Logout.

    ```bash
    exit
    ```

## 4. Configure the network <a name="configure"/>
----
**On each virtual machine** in your cluster, execute the following steps, one by one, line by line. *Do not copy-paste more than one line at a time*.

Note: When you replace the `<ip address>` field, <span style="color:red">**always use IP addresses, never use hostnames.**</span>

1. Login as `hadoop`. 

    ```bash
    ssh hadoop@<ip address>
    ```  
    
    If asked for a password, you did not complete steps 2 and/or 3 successfully. Go back and repeat.

2. Edit the `/etc/hosts` file with the following command, providing the `root` user password if requested:

    ```bash
    sudo nano /etc/hosts
    ```
    
    **Delete all its contents and replace them with the following, modified according to your virtual machines and hostnames selected at step 0**:

    ```
    127.0.0.1    localhost
    172.16.0.17  datanode1
    172.16.0.3   datanode2
    172.16.0.167 datanode3
    172.16.0.49  datanode4
    172.16.0.221 namenode
    ```

3. Edit the `/etc/hostname` file with the following command, providing the `root` user password if requested:

    ```bash
    sudo nano /etc/hostname
    ```

    **Replace its content with the hostname corresponding to the IP address of the virtual machine selected at step 0**.
    For example, on the machine with IP address 172.16.0.17, the `/etc/hostname` file should contain the following **single line**:
    ```
    datanode1
    ```

4. <span style="color:red">**IMPORTANT**</span>: you must make sure that the **name node has a password-less access to the data nodes**. Hence, <span style="color:red">on the **namenode** machine only</span>, run the following commands <span style="color:red">for every **datanode** machine</span>, providing the `hadoop` user password when requested:

    ```bash
    ssh-copy-id -i /home/hadoop/.ssh/id_rsa.pub hadoop@datanode1
    ssh-copy-id -i /home/hadoop/.ssh/id_rsa.pub hadoop@datanode2
    ssh-copy-id -i /home/hadoop/.ssh/id_rsa.pub hadoop@datanode3
    ssh-copy-id -i /home/hadoop/.ssh/id_rsa.pub hadoop@datanode4
    ``` 

5. Reboot the machine with the following command, providing the `root` user password if requested:

    ```bash
    sudo reboot
    ```

**From now on, we will never use again the root privileges, i.e. `sudo` commands.**
